# Analysis

```python
import json
import nltk
import re
```

### Load the script data

```python
filename = 'That70sData.json'
data = json.load(open(filename,'rb'))
```

```python
data.keys()
```

```python
data['8'][0]
```

## Combine scripts from all episodes

```python
all_scripts = ""
num_episodes = 0
for season in data:
    episodes = data[season]
    for episode in episodes:
        all_scripts += "".join(episode['script'])
        if episode['script']:
            num_episodes += 1
```

## Regular expressions for text concordance

```python
# pattern = r'in your ass'
pattern = r'your ass'
idxs = [(m.start(0), m.end(0)) for m in re.finditer(pattern, all_scripts)]

j = 50
print("Found {} results.\n{}\n".format(len(idxs),'-'*20))
for n,idx in enumerate(idxs):
    print("{}: ".format(n) + all_scripts[(idx[0]-j):(idx[1]+j)]+'\n')
```

## Do text concordance with NLTK

```python
tokens = nltk.word_tokenize(all_scripts)
text = nltk.Text(tokens)
```

```python
text.concordance('foot')
```

```python
text.concordance('dumbass')
```
